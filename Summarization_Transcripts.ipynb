```python
# @title Using pdftotext
import pdftotext
def text_extraction(path):
  with open(path, "rb") as f:
      pdf = pdftotext.PDF(f)
  text = []
  # All pages
  for txt in pdf:
    text.append(txt)
  return text
```

```python
import os
texts = []
def find_pdfs(folder_path):
    pdf_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith('.pdf'):
                pdf_files.append(os.path.join(root, file))
    return pdf_files

# Provide the path of the folder you want to search for PDF files
folder_path = '/content/drive/MyDrive/fin'

pdf_paths = find_pdfs(folder_path)
for pdf_path in pdf_paths:
    texts.append(text_extraction(pdf_path))
    print(pdf_path)
```

```python
# @title using glob
import glob

directory = '/content'
pattern = directory + '/*.pdf'

texts = []

file_list = glob.glob(pattern)

for file in file_list:
    texts += text_extraction(file)
```

**Function to Summarize**

```python
def summarize (model, text_list):
  summary = []
  for i, para in enumerate(text_list):

    completion = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "You are a business analyst with expertise in summarizing business meeting transcripts."},
            {"role": "user", "content": f'''summarize the following paragraph with
            short simple sentences. If you encounter any difficulty, simply say so and move on: "{para}"'''}
            ]
        )
    text = completion.choices[0].message.content
    summary.append(text)
    print (i, '/', len(text_list))

  return summary
```

**Function to identify key takeaways**

```python
def takeaways (model, text_list):

  takeaway_list = []
  full_summary = '\n\n'.join(text_list)

  completion = client.chat.completions.create(
      model=model,
      messages=[
          {"role": "system", "content": "You are a business analyst."},
          {"role": "user", "content": f'list out key takeaways from the following text: "{full_summary}"'}
          ]
  )
  takeaway_list.append(completion.choices[0].message.content)
  return (takeaway_list)
```

**Function to write to file**

```python
def write_to_file(file_path, summary_list):
  # Specify the file path
  file_path = file_path

  summarized_text = '\n\n'.join(summary_list)

  with open(file_path, 'w') as file:
    file.write(summarized_text)

  print(f'String has been saved to {file_path}')
```

```python
# @title Summarizing each page
summary_text = summarize ('gpt-3.5-turbo', texts)
write_to_file ('/content/drive/MyDrive/Colab Notebooks/Summarization/AshokLeyland_summary_Q12024_each_page_1', summary_text)
```

**Summarizing modified para_wise**

```python
import re
```

```python
def split_into_paragraphs(long_string):
    paragraphs = long_string.split("\n\n")
    return paragraphs
```

```python
def count_words(input_string):
    words = input_string.split()
    return len(words)
```

```python
for text,pdf_path in zip(texts,pdf_paths):
  names = namingFunc(pdf_path)
  textss = '\n'.join(text)
  paras = split_into_paragraphs(textss)
  para_list = []
  for i,item in enumerate (paras):
    if count_words(item)> 500:
      para_list.append(item)
    elif i == len(paras)-1:
      para_list.append(item)
    else:
      paras[i+1]= item + ' ' + paras[i+1]
  summary_para_list = summarize('gpt-3.5-turbo-16k', para_list)
  summary_path = f'/content/drive/MyDrive/fin/{names[0]}/{names[1]}/{names[2]}/summary/'
  if not os.path.exists(summary_path):
    os.makedirs(summary_path)
  write_to_file (summary_path+names[0]+'.txt', summary_para_list)

  takeaway_list = takeaways ('gpt-3.5-turbo-16k', summary_para_list)
  takeaway_path = f'/content/drive/MyDrive/fin/{names[0]}/{names[1]}/{names[2]}/keytakeaways/'
  if not os.path.exists(takeaway_path):
    os.makedirs(takeaway_path)
  write_to_file (takeaway_path+names[0]+'.txt', takeaway_list)
```
